# numpy.interp()

返回一階線性的填補值

```python
import numpy as np
import matplotlib.pyplot as plt

x = 2.5
xp = [1, 2, 3]
fp = [3, 2, 0]
y = np.interp(x, xp, fp) 
y
```

```python
plt.plot(xp, fp, '-o')
plt.plot(x, y, 'x')
plt.show()
```
